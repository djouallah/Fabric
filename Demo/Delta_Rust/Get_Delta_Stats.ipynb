```python
#the function read all the tables in the default Lakehouse and save it in a Table
import duckdb
from deltalake import DeltaTable
import os
from datetime import datetime
from deltalake.writer import write_deltalake
from trident_token_library_wrapper import PyTridentTokenLibrary
aadToken = PyTridentTokenLibrary.get_access_token("storage")
storage_options={"bearer_token": aadToken, "use_fabric_endpoint": "true"}
def OneLake(Delta_Path) :
  db = spark.conf.get("trident.lakehouse.name") 
  List_tables = [ f.name for f in os.scandir('/lakehouse/default/Tables/') if f.is_dir() ]
  duckdb.sql(' set enable_progress_bar = False')
  for tbl in List_tables :
    dt = DeltaTable('/lakehouse/default/Tables/'+ tbl)
    delta = ['/lakehouse/default/Tables/' +tbl + "/" + i for i in dt.files()]
    print(tbl)
    df_stats= duckdb.sql(f'''select *, '{tbl}' as table, '{db}' as db from parquet_metadata({delta}) ''').arrow()
    write_deltalake( Delta_Path ,df_stats ,mode='append',storage_options=storage_options)
  print('done')
OneLake("abfss://Delta_Table@onelake.dfs.fabric.microsoft.com/Delta_Table.Lakehouse/Tables/Delta_stat")
```
