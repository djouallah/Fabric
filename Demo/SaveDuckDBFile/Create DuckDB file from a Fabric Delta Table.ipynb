```python
from deltalake import DeltaTable
import os
import duckdb
from datetime import datetime
def Duck_Delta() :
  con=duckdb.connect('/lakehouse/default/Files/Data.duckdb')
  Workspace_Path = '/lakehouse/default/'
  List_tables = [ f.name for f in os.scandir(Workspace_Path +'Tables/') if f.is_dir() ]
  print(List_tables)
  con.sql(' set enable_progress_bar = False')
  for tbl in List_tables :
        dt = DeltaTable(Workspace_Path+'Tables/'+ tbl)
        delta = [Workspace_Path + 'Tables/' +tbl + "/" + i for i in dt.files()]
        print('Loading : ' +tbl)
        con.sql(f' create table IF NOT EXISTS {tbl} as select * from parquet_scan({delta})')
  con.sql('checkpoint')
  con.close()
Duck_Delta()
```
